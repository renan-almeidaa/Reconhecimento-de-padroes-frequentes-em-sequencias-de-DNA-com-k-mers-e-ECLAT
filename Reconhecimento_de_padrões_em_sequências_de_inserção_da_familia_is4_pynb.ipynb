### **K-mers**

k-mers são substrings de tamanho 𝐾 onde cada substring é chamada de k-mer, as
quais no contexto das macromoléculas, seriam nucleotídeos (A, T, G, e C) contidos em uma sequência biológica como DNA, RNA e proteínas. 

Por exemplo, considere a sequência
"AGCTTGATTGAC"contendo 12 nucleotídeos. Extraindo todos os k-mers (substrings) de
tamanho 4 dessa sequência, obtemos o resultado apresentado na figura abaixo:

k-mers-exemplo.jpg


Essa técnica é útil na mineração das sequências com o algoritmo de ECLAT, visto que
irá resultar em todas as possíveis substrings de cada sequência em análise gerando um conjunto
de transações diferentes, onde o algoritmo calculará o suporte de cada k-mer.

Quanto maior a frequência de um k-mer, mais forte é sua relevância, assim, sendo um
elemento em comum entre as sequências de inserção e determinante para a identificação da IS.
 

### **ECLAT**

ECLAT vem de Equivalence Class Clustering and Bottom-Up Lattice Traversal, ou em
português, Agrupamento de Classes de Equivalência e Travessia de rede de Baixo para Cima. É
um algoritmo de Machine Learning de mineração de regras de associação, sendo ele uma versão
[mais otimizada, eficiente e escalável do algoritmo Apriori](https://www.geeksforgeeks.org/ml-eclat-algorithm/).

Algoritmos de mineração de regras de associação são úteis para encontrar aspectos e
padrões comuns, extraindo regras e descobrindo relacionamentos entre a ocorrência de itens em
transações de uma base de dados.

Algoritmos de mineração são excelentes para encontrar associações entre dados e são
utilizados, por exemplo, pela [Netflix](https://www.cs.uic.edu/~liub/KDD-cup-2007/NetflixPrize-description.pdf), uma empresa provedora de filmes e séries via streaming,
para analisar os conteúdos que os clientes consomem e apresentar recomendações personalizadas,
melhorando assim, a experiência do usuário assinante.

Diferentemente da maioria dos outros algoritmos, o ECLAT **não gera** a confiança e o
lift, essenciais em outros modelos. Por esse motivo, há menos etapas no processamento de uma
base de dados, tornando sua execução [naturalmente mais rápida](https://towardsdatascience.com/the-eclat-algorithm-8ae3276d2d17)

## **Aplicando os métodos na base de dados sintética**



### **Preparação dos dados**

Vamos começar instalando as bibliotecas necessárias.

*   **pyEclat** é a biblioteca que contém a implementação do algoritmo de ECLAT que será utilizado para o reconhecimento de padrões das sequências. Mais informações em https://pypi.org/project/pyECLAT/.

```python
import pandas as pd
from Bio import SeqIO
from pyECLAT import ECLAT
import csv
import json
```

Utilizando o import SeqIO, fazemos a extração das sequências contidas no arquivo .fasta em uma lista, para melhor manipulação.

```python
seq_list = []
lst =[]
seq_list = [str(s.seq) for s in SeqIO.parse(open('base_sintetica.fasta'), 'fasta')]
  
seq_list
```

Criando um dataframe com as sequências acima

```python
df_seqs = pd.DataFrame(seq_list, columns = ['sequencias'])
```

#### **Aplicando a técnica de k-mer**

No código abaixo, vamos preparar as sequências para a utilização no algoritmo de ECLAT aplicando a técnica de K-mer.

```python
#O parâmetro size será o tamanho K dos k-mers
def getKmers(sequence, size=4):
    return [sequence[x:x+size].lower() for x in range(len(sequence) - size + 1)]

df_seqs['sequencias'] = df_seqs.apply(lambda x: getKmers(x['sequencias']), axis=1)

df_seqs
```

Observe que na representação do dataframe "df_seqs" acima há apenas uma coluna contendo uma lista com todos os k-mers.

É necessário que cada k-mer tenha sua própria coluna, para que o ECLAT consiga fazer suas operações. Para isso, aplicaremos o método "pd.Series". Esse método gerará um dataframe com todos os elementos das listas.

OBS: Note que as linhas do dataframe possui listas de tamanhos diferentes. O pd.Series resolverá esse problema criando um dataframe com o comprimento máximo da lista e preencherá os valores ausentes com NaN

```python
df_seqs = df_seqs['sequencias'].apply(pd.Series)
```

E pronto, temos nossa base de dados sintética formada e apta para ser utilizada.

```python
df_seqs
```

#### **ECLAT**

E finalmente, vamos encaixar os dados no algoritmo de ECLAT e testar a sua eficiência.

```python
eclat = ECLAT(data = df_seqs)
```

O método .uniq_ apresenta uma lista com todas as sequências únicas, sem repetição. Com a aplicação de k-mers, foi possível gerar 249 sequências únicas.

```python
len(eclat.uniq_)
```

Encaixando os dados no algoritmo de ECLAT, utilizei os seguintes parâmetros:


*   suporte minimo de 0.6, 
*   mínimo de 1 combinação
*   máximo 2 combinações

Por
conta da característica do k-mers onde as substrings são deslizantes, é possível encontrar motifs
de tamanho até 5 em itemsets que contém 2 itens combinados, já que as sequências foram
transformadas em k-mers de tamanho 4.

```python
indices, suporte = eclat.fit(min_support=0.6, min_combination=1, max_combination=2)
```

Ordenando resultado pelo suporte.

```python
suporte_ordenado = {k: v for k, v in sorted(suporte.items(), key=lambda item: item[1], reverse=True)}
```

```python
suporte_ordenado
```
